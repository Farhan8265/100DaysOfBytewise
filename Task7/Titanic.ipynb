```
import pandas as pd
import matplotlib.pyplot as plt
# Load the dataset
titanic = pd.read_csv('titanic.csv')
ages = titanic['Age'].dropna().sort_values().reset_index(drop=True)
plt.figure(figsize=(10, 6))
plt.plot(ages, label='Passenger Ages')
plt.xlabel('Passenger Index')
plt.ylabel('Age')
plt.title('Trend of Passenger Ages over Time')
plt.legend()
plt.show()
```

```
# Exercise 2: Bar Chart using Matplotlib
pclass_counts = titanic['Pclass'].value_counts()
# Create the bar chart
plt.figure(figsize=(10, 6))
pclass_counts.plot(kind='bar')
plt.xlabel('Passenger Class')
plt.ylabel('Frequency')
plt.title('Frequency of Different Passenger Classes')
plt.show()
```

```
# Exercise 3: Scatter Plot using Matplotlib
# Relationship between Age and Fare
plt.figure(figsize=(10, 6))
plt.scatter(titanic['Age'], titanic['Fare'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs. Fare')
plt.show()
```

```
# Exercise 4: Pairplot using Seaborn
import seaborn as sns
sns.pairplot(titanic[['Age', 'Fare', 'Pclass', 'Survived']].dropna())
plt.show()
```

```
# Exercise 5: Box Plot using Seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Age', data=titanic)
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.title('Box Plot of Age Distribution across Passenger Classes')
plt.show()
```

```
# Exercise 6: Heatmap using Seaborn
# Correlation matrix
corr_matrix = titanic[['Age', 'Fare', 'Pclass', 'Survived']].corr()

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```
# Exercise 7: Subplot Grid using Matplotlib
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Line plot
axs[0, 0].plot(ages)
axs[0, 0].set_title('Line Plot of Passenger Ages')

# Bar chart
pclass_counts.plot(kind='bar', ax=axs[0, 1])
axs[0, 1].set_title('Bar Chart of Passenger Classes')

# Scatter plot
axs[1, 0].scatter(titanic['Age'], titanic['Fare'], alpha=0.5)
axs[1, 0].set_title('Scatter Plot of Age vs. Fare')

# Box plot
sns.boxplot(x='Pclass', y='Age', data=titanic, ax=axs[1, 1])
axs[1, 1].set_title('Box Plot of Age Distribution')

plt.tight_layout()
plt.show()
```

```
# Exercise 8: Customize Seaborn Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Age', data=titanic, palette='pastel')
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.title('Customized Box Plot of Age Distribution across Passenger Classes', fontsize=14)
plt.show()
```

```
# Exercise 9: Violin Plot using Seaborn
plt.figure(figsize=(10, 6))
sns.violinplot(x='Pclass', y='Age', data=titanic)
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.title('Violin Plot of Age Distribution across Passenger Classes')
plt.show()
```

```
# Exercise 10: Complex Visualization with Matplotlib and Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(titanic['Age'].dropna(), kde=True, color='blue', bins=30)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram with KDE Overlay of Passenger Ages')
plt.show()
```
