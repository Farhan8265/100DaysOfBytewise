```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Load the Iris dataset
iris = sns.load_dataset('iris')
```

```
# Exercise: Calculate Mean, Median, and Mode of the Sepal Lengths in Iris Dataset
mean_sepal_length = iris['sepal_length'].mean()
median_sepal_length = iris['sepal_length'].median()
mode_sepal_length = iris['sepal_length'].mode()[0]

print("Mean Sepal Length: ", mean_sepal_length)
print("Median Sepal Length: ", median_sepal_length)
print("Mode Sepal Length: ", mode_sepal_length)
```

```
# Exercise: Calculate Variance and Standard Deviation of Petal Widths in Iris Dataset
variance_petal_width = iris['petal_width'].var()
std_petal_width = iris['petal_width'].std()

print("Variance of Petal Width: ", variance_petal_width)
print("Standard Deviation of Petal Width: ", std_petal_width)
```

```
# Exercise: Define a Random Variable for Sepal Length and Calculate Probability Distribution.
sepal_length_distribution = iris['sepal_length'].value_counts(normalize=True).sort_index()
sepal_length_distribution
```

```
# Exercise: Plot Probability Distribution of Sepal Lengths Using a Histogram.
plt.figure(figsize=(10, 6))
sns.histplot(iris['sepal_length'], kde=False, stat='probability', bins=30)
plt.xlabel('Sepal Length')
plt.ylabel('Probability')
plt.title('Probability Distribution of Sepal Lengths')
plt.show()
```

```
# Exercise: Calculate Cumulative Distribution Function (CDF) for Petal Lengths and Plot it
plt.figure(figsize=(10, 6))
sns.ecdfplot(iris['petal_length'])
plt.xlabel('Petal Length')
plt.ylabel('CDF')
plt.title('Cumulative Distribution Function of Petal Lengths')
plt.show()
```

```
# Exercise: Calculate and Plot the Probability Density Function (PDF) for Sepal Width
plt.figure(figsize=(10, 6))
sns.kdeplot(iris['sepal_width'])
plt.xlabel('Sepal Width')
plt.ylabel('Density')
plt.title('Probability Density Function of Sepal Widths')
plt.show()
```

```
# Exercise: Determine Probability of Randomly Selected Iris Flower Having a Petal Length Greater than  Given Value
given_value = 5.0
prob_greater_than_value = (iris['petal_length'] > given_value).mean()
print("Probability of petal length greater than", given_value, ":", prob_greater_than_value)
```

```
# Exercise: Perform Hypothesis Test
# Hypothesis test between setosa and versicolor
setosa_petal_length = iris[iris['species'] == 'setosa']['petal_length']
versicolor_petal_length = iris[iris['species'] == 'versicolor']['petal_length']

# Perform an independent t-test
t_stat, p_value = stats.ttest_ind(setosa_petal_length, versicolor_petal_length)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```
# Exercise: Calculate and Interpret the Covariance and Correlation Between Sepal Length and Sepal Width
# Covariance
covariance_sepal_length_width = iris[['sepal_length', 'sepal_width']].cov().iloc[0, 1]

# Correlation
correlation_sepal_length_width = iris[['sepal_length', 'sepal_width']].corr().iloc[0, 1]

print("Covariance between Sepal Length & Sepal Width: ", covariance_sepal_length_width)
print("Correlation between Sepal Length & Sepal Width: ", correlation_sepal_length_width)
```
