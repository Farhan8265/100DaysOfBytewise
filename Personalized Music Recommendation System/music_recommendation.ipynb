```python
import pandas as pd
import re
```

```python
spotify = pd.read_csv('spotify.csv')
spotify.head()
```

```python
spotify.info()
```

```python
spotify.shape
```

```python
# Handle missing values
spotify.fillna("N/A", inplace=True)
```

```python
def clean_tracks(name):
    name = re.sub(r'[^\w\s]', '', name)
    name = re.sub(r'\s+', ' ', name)
    name = name.strip()
    return name
```

```python
spotify['track_name'] = spotify['track_name'].apply(clean_tracks)
```

```python
spotify['explicit'] = spotify['explicit'].astype(int)
```

```python
spotify.head(5)
```

```python
spotify['track_name'] = spotify['track_name'].replace('', 'Unknown Track')
spotify['album_name'] = spotify['album_name'].replace('', 'Unknown Album')
spotify['artists'] = spotify['artists'].replace('', 'Unknown Artist')
```

```python
# Sort data
spotify = spotify.sort_values(by=['track_name']).reset_index(drop=True)
```

```python
spotify.head()
```

```python
# Remove duplicates
spotify.drop_duplicates(subset=['track_name', 'artists'], keep='first', inplace=True)
```

```python
spotify = spotify.drop(columns=['Unnamed: 0'])
```

```python
spotify.head()
```

```python
# Converting into lower case
spotify['track_name'] = spotify['track_name'].str.lower()
spotify['artists'] = spotify['artists'].str.lower()
spotify['album_name'] = spotify['album_name'].str.lower()
```

```python
spotify.head()
```

```python
spotify['combined_features'] = spotify['track_name'] + " " + spotify['artists'] + " " + spotify['album_name'] + " " + spotify['track_genre']
```

```python
new_spotify = spotify[ ['track_name', 'artists', 'album_name', 'track_genre', 'combined_features'] ]
```

```python
new_spotify.head()
```

```python
new_spotify.shape
```

```python
new_spotify.insert(0, 'ID', range(1, len(new_spotify) + 1))
```

```python
new_spotify.head(10)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
```

```python
# Vectorizing the text (using TF-IDF)
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(new_spotify['combined_features'])
```

```python
# Fit Nearest Neighbors model using cosine similarity metric
knn_model = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=6, n_jobs=-1)
knn_model.fit(tfidf_matrix)
```

```python
# Function to recommend songs using KNN
def recommend_songs(song_name, knn_model, df, tfidf_matrix):
    song_name = song_name.lower()

    # Check if the song exists in the dataset
    if song_name not in new_spotify['track_name'].values:
        print(f"Sorry, '{song_name}' not found in the dataset.")
        return

    try:
        # Find the index of the song in the dataset
        idx = new_spotify[new_spotify['track_name'] == song_name].index[0]

        # Reshape the TF-IDF vector for the input song to be used for similarity comparison
        song_vector = tfidf_matrix[idx].reshape(1, -1)

        # Find most similar songs
        distances, indices = knn_model.kneighbors(song_vector)

        # Get the indices of the most similar songs
        song_indices = indices.flatten()[1:6]
        if len(song_indices) == 0:
            return "No recommendations available."

        # Ensure all indices are within the range of the DataFrame
        if any(index >= len(new_spotify) or index < 0 for index in song_indices):
            return "Index out of Range Error. Please check the Dataset or Input song."

        # Return the top 5 recommended songs
        return new_spotify.iloc[song_indices][['track_name', 'artists']]

    except IndexError:
        return "Index error encountered. Please check the Dataset or Input song."
```

```python
import pickle
with open("music_recommendation.pkl", "wb") as f:
    pickle.dump(knn_model, f)
with open("tfidf_vectorizer.pkl", "wb") as f:
    pickle.dump(tfidf, f)
```
